## Chuẩn bị thêm dữ liệu từ ảnh chụp camera

```python
import cv2
import os

label = "Mask_Nose_Mouth"
```

```python
# Ảnh train
train_dir = "./dataset/train/"
label_train_dir = os.path.join(train_dir, label)

cap = cv2.VideoCapture(0)

i = 0
while True: 
    ret, frame = cap.read()
    if not ret: 
        break
        
    cv2.imshow("Image", frame)
    
    i += 1
    print(f"Iterator: {i}")
    if i >= 60 and i <= 660: 
        image_name = f"Cap_{label}_{i:05.0f}.jpg"
        cv2.imwrite(os.path.join(label_train_dir, image_name), frame)
        
    if cv2.waitKey(1) == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
